```python
# 최장연체일수 컬럼 분포 시각화

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 한글 폰트 설정 (환경에 따라 다를 수 있음)
plt.rcParams['font.family'] = 'Malgun Gothic' 
plt.rcParams['axes.unicode_minus'] = False

# 데이터 준비 (0일인 그룹과 연체 경험이 있는 그룹 분리)
df = temp_card_2103
delq_only = df[df['PYE_MAX_DLQ_DAY'] > 0]

# 1. 전체 분포와 연체 경험자 분포 비교 (Dual Plot)
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 왼쪽: 전체 분포 (Log Scale 적용)
sns.histplot(df['PYE_MAX_DLQ_DAY'], bins=50, ax=axes[0], kde=True, color='gray')
axes[0].set_yscale('log') # 0일 데이터가 너무 많으므로 로그 스케일 적용
axes[0].set_title('전체 연체 일수 분포 (Log Scale)', fontsize=14)
axes[0].set_xlabel('연체 일수')
axes[0].set_ylabel('고객 수 (Log)')

# 오른쪽: 연체 경험자(>0)들만 상세 분석
sns.histplot(delq_only['PYE_MAX_DLQ_DAY'], bins=50, ax=axes[1], kde=True, color='crimson')
axes[1].set_title('연체 경험자(>0) 내 상세 분포', fontsize=14)
axes[1].set_xlabel('연체 일수')
axes[1].set_ylabel('고객 수')

# 2. 누적 분포도 (ECDF) - "상위 몇 %가 며칠 연체인가?" 파악용
plt.figure(figsize=(10, 5))
sns.ecdfplot(data=df, x='PYE_MAX_DLQ_DAY', color='blue')
plt.axvline(x=30, color='red', linestyle='--', label='산업 표준(30일)')
plt.title('연체 일수 누적 분포 (ECDF)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()

# 3. 박스 플롯 (Outlier 확인)
plt.figure(figsize=(10, 4))
sns.boxplot(x=df['PYE_MAX_DLQ_DAY'], color='orange')
plt.title('연체 일수 박스 플롯 (이상치 확인)', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
temp_card_2103.sort_values('CUST_ID').reset_index(drop=True) # 일단 정렬부터 맞추기
temp_card_2106.sort_values('CUST_ID').reset_index(drop=True)

is_equals = temp_card_2103.equals(temp_card_2106)

if not is_equals: # 두 테이블이 완전히 같지 않으면
    diff_mask = (temp_card_2103 != temp_card_2106).any(axis=1) # axis=1 : 옆으로 훑어보기
    mismatch_data = temp_card_2103[diff_mask]
    print(len(mismatch_data))
```

```python
diff_table = (temp_card_2103 != temp_card_2106)
mismatch_cnt = diff_table.sum(axis=0).reset_index(name='cnt') # 각 컬럼에서 세로로 훑어보면서 몇개의 True가 있는지 sum
mismatch_cnt[(mismatch_cnt['cnt']>0) & (mismatch_cnt['cnt']<300000)].sort_values(by='cnt', ascending=False).head(20)
```
