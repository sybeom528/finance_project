```python
import pandas as pd
import gc

def find_special_patterns(df12, df9, df6, df3, id_col='CUST_ID'):

    print("데이터 인덱스 정렬 및 교집합 추출")
    
    idx12 = df12.set_index(id_col)
    idx9 = df9.set_index(id_col)
    idx6 = df6.set_index(id_col)
    idx3 = df3.set_index(id_col)

    common_ids = idx12.index.intersection(idx9.index).intersection(idx6.index).intersection(idx3.index)
    common_cols = idx12.columns.intersection(idx9.columns).intersection(idx6.columns).intersection(idx3.columns)

    m12 = idx12.loc[common_ids, common_cols]
    m9 = idx9.loc[common_ids, common_cols]
    m6 = idx6.loc[common_ids, common_cols]
    m3 = idx3.loc[common_ids, common_cols]

    print(f"✅ 매칭 완료: {len(common_ids)}명 / 검사 대상: {len(common_cols)}개 컬럼\n")

    def check_same(past, recent):
        return (past == recent) | (past.isna() & recent.isna())

    is_all_same = check_same(m12, m9) & check_same(m9, m6) & check_same(m6, m3)
    static_mask = is_all_same.all()
    static_columns = static_mask[static_mask].index.tolist()
    
    num_cols = m12.select_dtypes(include='number').columns
    n12, n9, n6, n3 = m12[num_cols], m9[num_cols], m6[num_cols], m3[num_cols]

    is_increasing = (n12 <= n9) & (n9 <= n6) & (n6 <= n3)
    inc_mask = is_increasing.all()
    
    increasing_columns = [col for col in inc_mask[inc_mask].index.tolist() if col not in static_columns]

    print(f"미변화 컬럼: {len(static_columns)}개")
    print(f"누적/증가 컬럼: {len(increasing_columns)}개")

    del idx12, idx9, idx6, idx3, m12, m9, m6, m3
    del is_all_same, n12, n9, n6, n3, is_increasing
    gc.collect()

    return {
        'static': static_columns,
        'increasing': increasing_columns
    }

result = find_special_patterns(df_202103, df_202106, df_202109, df_202112)
print("안 변하는 변수들:", result['static'])
print("증가만 하는 변수들:", result['increasing'])
```
