```python
def parse_v2tone(tone_str):
    """V2Tone 필드 → 7개 차원으로 분해"""
    if pd.isna(tone_str) or tone_str == '':
        return {
            'tone': None, 'pos_score': None, 'neg_score': None,
            'polarity': None, 'activity': None, 'self_ref': None, 'word_count': None
        }
    vals = tone_str.split(',')
    return {
        'tone':       float(vals[0]) if len(vals) > 0 else None,
        'pos_score':  float(vals[1]) if len(vals) > 1 else None,
        'neg_score':  float(vals[2]) if len(vals) > 2 else None,
        'polarity':   float(vals[3]) if len(vals) > 3 else None,
        'activity':   float(vals[4]) if len(vals) > 4 else None,
        'self_ref':   float(vals[5]) if len(vals) > 5 else None,
        'word_count': int(float(vals[6])) if len(vals) > 6 else None,
    }

# DataFrame에 적용
tone_df = gkg['V2Tone'].apply(parse_v2tone).apply(pd.Series)
gkg = pd.concat([gkg, tone_df], axis=1)
display(tone_df)
```

```python
import gdelt
import pandas as pd

pd.set_option('display.max_columns', 30)
pd.set_option('display.width', 160)

gd = gdelt.gdelt(version=2)
```

```python
TARGET = ['2025 04 14']  # 원하는 날짜

print("[1/3] Events…")
events   = gd.Search(TARGET, table='events',   coverage=True, output='df')
print("[2/3] Mentions…")
mentions = gd.Search(TARGET, table='mentions', coverage=True, output='df')
print("[3/3] GKG…")
gkg      = gd.Search(TARGET, table='gkg',      coverage=True, output='df')

print(f"\nshape → events={events.shape}, mentions={mentions.shape}, gkg={gkg.shape}")
```

```python
def find_col(df, keyword):
    """대소문자·구분자(. _) 무시 부분일치 검색."""
    norm = lambda s: s.lower().replace('.', '').replace('_', '')
    target = norm(keyword)
    for c in df.columns:
        if target in norm(c):
            return c
    raise KeyError(f"'{keyword}' 없음. 컬럼: {list(df.columns)}")

gkg_url_col = find_col(gkg, 'DocumentIdentifier')
print(f"GKG URL 컬럼: '{gkg_url_col}'")
```

```python
# (a) Events ⨝ Mentions
ev_mn = events.merge(
    mentions, on='GLOBALEVENTID', how='inner', suffixes=('_evt', '_mnt')
)
print(f"[JOIN 1] Events ⨝ Mentions : {len(ev_mn):,} rows")

# (b) ⨝ GKG on URL
joined = ev_mn.merge(
    gkg, left_on='MentionIdentifier', right_on=gkg_url_col,
    how='inner', suffixes=('', '_gkg')
)
print(f"[JOIN 2] ⨝ GKG             : {len(joined):,} rows × {joined.shape[1]} cols")
```

```python
stages = pd.DataFrame({
    'stage': ['Events (raw)', 'Mentions (raw)', 'GKG (raw)',
              'Events ⨝ Mentions', '⨝ GKG (final)'],
    'rows':  [len(events), len(mentions), len(gkg), len(ev_mn), len(joined)],
})
stages
```

```python
matched = joined['GLOBALEVENTID'].nunique()
print(f"Events {len(events):,}건 중 GKG까지 연결된 고유 이벤트: "
      f"{matched:,}건 ({matched/len(events)*100:.1f}%)")
print("→ 단일 15분 파일 기준 30~50%가 일반적. 낮으면 시간 윈도우 확장 필요")
```

```python
key_cols = [
    'GLOBALEVENTID', 'SQLDATE', 'EventCode', 'GoldsteinScale', 'AvgTone',
    'Actor1CountryCode', 'ActionGeo_CountryCode',
    'MentionSourceName', 'Confidence', 'MentionDocTone',
]
for kw in ['EnhancedThemes', 'V1.5Tone', 'AllNames']:
    try:    key_cols.append(find_col(joined, kw))
    except KeyError: pass

available = [c for c in key_cols if c in joined.columns]
joined[available].head(5)
```
